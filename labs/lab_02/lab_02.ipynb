## Task 2: Grouping and sorting data

```python
grouped_confirmed = df_confirmed.groupby('Country/Region').sum()
```

```python
grouped_confirmed.head(10)
```

```python
sorted_confirmed = grouped_confirmed.sort_values(by=grouped_confirmed.columns[-1], ascending=False)
sorted_confirmed.head(10)
```

```python
# Plotting the data
import matplotlib.pyplot as plt
import seaborn as sns

last_col = df_confirmed.iloc[-1]
last_day = last_col.index[-1]
plt.figure(figsize=(12,8))
plt.title("Top 10 countries with highest cases", fontsize=14)
sns.barplot(x = sorted_confirmed[last_day].head(10), y = sorted_confirmed[last_day].index[:10], orient='h')
plt.xlabel("Total cases by " + last_day)
plt.grid()
plt.show()
```

```python
## Time series analysis
def get_total_confirmed_world():
    total = df_confirmed.iloc[:,4:].apply(sum, axis=0)
    total.index = pd.to_datetime(total.index, format="%m/%d/%y")
    return total

print(get_total_confirmed_world())
```
